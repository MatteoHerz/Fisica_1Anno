### Esercizio 1
__Fate il plot delle due funzioni $f_1(x) = (x^2-x+2)/(x^4+2)$ in $\,0\,\le x \le \,4$ e $f_2(x) = \cos(x+1)/(x^2+ 1)$ in $\,-2\,\le x \le \,2$ nello stesso grafico. Inserire un titolo per il plot, i titoli degli assi e le legende.__

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 5))

x1= np.linspace(0,4,100)
x2= np.linspace(-2,2,100)
y1= (x1**2-x1+2)/(x1**4+2)
y2= np.cos(x2+1)/(x2**2+1)

ax.set_xlabel("x coordinate")
ax.set_ylabel("f(x)")
ax.set_title("Esercizio 1")

ax.grid()
ax.plot(x1,y1, label="f1(x)=(x2âˆ’x+2)/(x4+2)", c='r')
ax.plot(x2,y2, label="f2(x)=cos(x+1)/(x2+1)", c='b');
ax.legend(fontsize="12");
````

### Esercizio 2
__Generate le coordinate x e y di 100 punti casuali uniformemente distribuiti nel rettangolo di vertici (0,0) (0,5) (8,0) (8,5).__<BR>
__Fate il plot dei punti generati.__<BR>
__Calcolate il valor medio della distanza dei punti dall'origine.__<BR>


````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng

rng=default_rng()
x = rng.uniform(0.,high=8.,size=100)
y = rng.uniform(0.,high=5.,size=100)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 5))
ax.scatter(x,y, c="r");
ax.set_title("Esercizio 2")

mediadistanza = np.mean(np.sqrt(x**2+y**2))

print(f"La media della distanza dei punti dall'origine: {mediadistanza}")
````

### Esercizio 3
__Fate il plot delle funzioni $f_n(x) = (\sin(x^2-x+2))^n/(x^4+2)$ con $n$ = 1,2,3,4 in quattro plot disposti su due righe.Inserire i titoli degli assi e le legende. Salvate il plot su disco in formato pdf.__

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 7))
x = np.linspace(-5,5,150)
y1= np.sin(x**2-x+2)/(x**4+2)
y2= (np.sin(x**2-x+2))**2/(x**4+2)
y3= (np.sin(x**2-x+2))**3/(x**4+2)
y4= (np.sin(x**2-x+2))**4/(x**4+2)

ax[0,0].grid()
ax[0,1].grid()
ax[1,0].grid()
ax[1,1].grid()

ax[0,0].set_xlabel("x coordinate")
ax[0,1].set_xlabel("x coordinate")
ax[1,0].set_xlabel("x coordinate")
ax[1,1].set_xlabel("x coordinate")

ax[0,0].set_ylabel("f(x)")
ax[0,1].set_ylabel("f(x)")
ax[1,0].set_ylabel("f(x)")
ax[1,1].set_ylabel("f(x)")

ax[0,0].plot(x,y1, c='b', label="sin(x^2-x+2)/(x^4+2)")
ax[0,1].plot(x,y2, c='r', label="sin(x^2-x+2))^2/(x^4+2)")
ax[1,0].plot(x,y3, c='y', label="sin(x^2-x+2))^3/(x^4+2)")
ax[1,1].plot(x,y4, c='g', label="sin(x^2-x+2))^4/(x^4+2)");

ax[0,0].legend(fontsize="8");
ax[0,1].legend(fontsize="7");
ax[1,0].legend(fontsize="8");
ax[1,1].legend(fontsize="7");

fig.savefig('plot.pdf')
````

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 7))
x = np.linspace(-5,5,150)

ax = (ax1,ax2,ax3,ax4)
colori = ("r","purple","g","navy")

for i in ax:
    y= np.sin(x**2-x+2)**(ax.index(i)+1)/(x**4+2)
    i.grid()
    i.set_xlabel("x coordinate")
    i.set_ylabel(f"f(x) - {ax.index(i)+1}")
    i.plot(x,y, c=colori[ax.index(i)], label=f"sin(x^2-x+2)^{ax.index(i)+1}/(x^4+2)")
    i.legend(fontsize="7")
    
fig.savefig('plot.pdf')
````
