### Esercizio 1
__Fate un unico plot delle due funzioni $f_1(x) = \sin^2(x)/(x^2 +1)$ e $f_2(x) = \sin^2(2\,x -1)/(3\,x^3 +1)$ in 
$\,0\,\le x \le \,5$. Inserite il nome degli assi e le legende.__

````python
import numpy as np
import math as m
import matplotlib.pyplot as plt
t = np.arange(0., 5., 0.01)

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(t,(np.sin(t)**2/(2*t**2+1)),c='r', label= "sin(x)**2/(2*x**2+1)")
ax.plot(t,(np.sin(2*t-1)**2/(3*t**3+1)),c='b', label= "sin(2*x-1)**2/(3*x**3+1)")
ax.grid()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
````

### Esercizio 2
__La posizione, al tempo t, di un corpo che si muova sotto effetto della gravità terrestre è data da:__
$$y(t) = y_0 + v_{y0} t − \frac{1}{2}\, g\, t^2,\qquad x(t) = x_0 +v_{x0} t$$ 
__dove $g = 9.81 m/s^2$ e $x_0,\,y_0,\,v_{x0},\, v_{y0}$ sono le posizioni e velocità a $t = 0$. 
Supponiamo $y_0 = 100\, m,\,x_0 = 0,\, v_{y0} = 0, v_{x0} = 5\, m/s$. Disegnate la traiettoria $y = y(x)$ del corpo per $y > 0$.
Suggerimento: calcolate come prima cosa il tempo a cui il corpo arriva a $y = 0$.__

````python
import numpy as np
import math as m
import matplotlib.pyplot as plt

x0=0
vy0=0
vx0=5
g=9.81
y0=100
t = np.arange(0.,(y0*2/g)**0.5, 0.001)

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot((x0+t*vx0),(y0+vy0*t-0.5*g*t**2),c='navy')
ax.set_xlabel("x coordianate")
ax.set_ylabel("f(x)")
ax.set_title("Esercizio 2"); 
ax.grid()
````

### Esercizio 3
__Producete un campione casuale di 100 eventi estratti da una distribuzione normale, con valor medio $\mu = 1.7.$ e deviazione standard $\sigma = 0.4$. Calcolate il valor medio e la deviazione standard del campione. Fate l'istogramma delle frequenze del campione.__

````python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from numpy.random import default_rng

rng = default_rng()

A = rng.normal(loc=1.7, scale=0.4, size=100)
nbins = 100
xrange = (0,5)    # ntupla
fig, ax = plt.subplots()
nevent, bins, patches = ax.hist(A, nbins, range=xrange)

media = 1.7
deviazione_standard = 0.4

x = np.linspace(0, 5, 1000) 
y = (1 / (deviazione_standard * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media) / deviazione_standard) ** 2)


ax.set_xlabel('Valore')
ax.set_ylabel('Densità di probabilità')
ax.set_title('Distribuzione Gaussiana')
plt.grid()
ax.plot(x,y);
````

### Esercizio 4
__Costruite una funzione `f(n)` che generi due array, `x` e `y`, ciascuno di `n `numeri casuali distribuiti uniformemente nell'intervallo fra zero e due. `f` deve ordinare i due array in ordine crescente e fare il grafico usando i due array come coordinate dei punti da plottare.__

````python
def graf(n):
    x= np.random.uniform(0,2,n)
    y= np.random.uniform(0,2,n)
    X=np.sort(x)
    Y=np.sort(y)
    fig1, ax = plt.subplots(figsize=(8, 5))
    return ax.plot(X,Y)
print(graf(200))   
````
